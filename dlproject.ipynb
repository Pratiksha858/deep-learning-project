# feature engineering 
##we cant use datetime columns as feature for prediction so we calculate diffrance and create new column

```python
# Step 1: Convert date columns to datetime format
df['orighiredate_key'] = pd.to_datetime(df['orighiredate_key'])
df['terminationdate_key'] = pd.to_datetime(df['terminationdate_key'])

# Step 2: Calculate the difference and create a new column
df['employment_duration'] = (df['terminationdate_key'] - df['orighiredate_key']).dt.days

# Display the updated DataFrame with the new column
df
```

```python
# Step 1: Convert date columns to datetime format
df['birthdate_key'] = pd.to_datetime(df['birthdate_key'])
df['orighiredate_key'] = pd.to_datetime(df['orighiredate_key'])


# Step 2: Calculate the difference and create a new column
df['employ_age_at_hire'] = (df['orighiredate_key'] - df['birthdate_key']).dt.days

# Display the updated DataFrame with the new column
df
```

```python
# Step 1: Convert date columns to datetime format
df['recorddate_key'] = pd.to_datetime(df['recorddate_key'])
df['orighiredate_key'] = pd.to_datetime(df['orighiredate_key'])

# Step 2: Calculate the difference and create a new column
df['days_since_hire'] = (df['recorddate_key'] - df['orighiredate_key']).dt.days

# Display the updated DataFrame with the new column
df
```

```python
#BUSINESS_UNIT and Employee id not neccessary for stetus prediction
df = df.drop(columns=['BUSINESS_UNIT','recorddate_key','birthdate_key','gender_full','orighiredate_key','terminationdate_key','EmployeeID'])
```

```python
from sklearn.preprocessing import LabelEncoder
columns1=['city_name','department_name','job_title', 'gender_short','termreason_desc', 'termtype_desc','STATUS']
label_encoder = LabelEncoder()
for i in columns1:
    df[i] = label_encoder.fit_transform(df[i])
```

```python
def find_outliers_zscore(df, threshold=3):
    outliers = []
    for column in df.columns:
        z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
        column_outliers = df[ z_scores > threshold ][column]
        outliers.extend(column_outliers.index.tolist())
    
    return list(set(outliers))

outlier_indices = find_outliers_zscore(df)

# Visualize outliers using box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of the Dataset")
plt.xticks(rotation=45)
plt.show()
```

```python
def remove_outliers_zscore(df, threshold=3):
    outliers = set()
    for column in df.columns:
        z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
        outliers.update(df[z_scores > threshold].index.tolist())

    # Remove outliers from the dataset
    data_cleaned = df.drop(index=outliers)
    return data_cleaned

data_cleaned = remove_outliers_zscore(df)

# Visualize data_cleaned using box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_cleaned)
plt.title("Box Plot of Cleaned Dataset (Outliers Removed)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop the target variable from the DataFrame
X = df.drop(columns=['STATUS'])
y = df['STATUS']
X
```

```python
#  Normalize the features using Min-Max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
y
```

```python
#data is imbalance so smote is use
# Apply SMOTE to oversample the minority class

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train, y_train= smote.fit_resample(X, y)
```

```python
#Create and train the ANN model
from keras.models import Sequential
from keras.layers import Dense

# Initialize the ANN model
model = Sequential()

# Add the input layer and the first hidden layer
model.add(Dense(units=64, activation='relu', input_dim=X_scaled.shape[1]))

# Add more hidden layers (you can experiment with different architectures)
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))

# Add the output layer with a sigmoid activation function (for binary classification)
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train the model on the training data
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)
```

# evaluation

```python
# Evaluate the model's performance
loss, accuracy = model.evaluate(X_scaled, y)
print(f"Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}")
```

```python
#  Set the threshold to convert probabilities to binary predictions
threshold = 0.5
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > threshold).astype(int)
```

```python
#  Evaluate the model using classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_binary))
```
